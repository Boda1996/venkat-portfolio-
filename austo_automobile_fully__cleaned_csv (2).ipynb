## Data analysis

### Subtask:
Analyze the data to identify potential relationships between variables and detect outliers.


**Reasoning**:
Calculate the correlation matrix for numerical features, visualize it as a heatmap, and then analyze the relationship between categorical features and price using box plots.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the correlation matrix for numerical features
numeric_cols = ['Age', 'No_of_Dependents', 'Salary', 'Partner_salary', 'Total_salary', 'Price']
correlation_matrix = df[numeric_cols].corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# Analyze the relationship between categorical features and price
categorical_cols = ['Gender', 'Profession', 'Marital_status', 'Education', 'Personal_loan', 'House_loan', 'Partner_working', 'Make']
plt.figure(figsize=(20, 15))
for i, col in enumerate(categorical_cols):
    plt.subplot(4, 2, i + 1)
    sns.boxplot(x=col, y='Price', data=df)
    plt.title(f'Price vs. {col}')
    plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Reasoning**:
Analyze potential outliers in numerical features using box plots and scatter plots.  Document any notable patterns or relationships.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Analyze potential outliers in numerical features
numeric_cols = ['Age', 'No_of_Dependents', 'Salary', 'Partner_salary', 'Total_salary', 'Price']
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols):
    plt.subplot(3, 2, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Box plot of {col}')
plt.tight_layout()
plt.show()

# Further investigation using scatter plots
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols):
    if col != 'Price':  # Avoid plotting Price vs. Price
        plt.subplot(3, 2, i + 1)
        sns.scatterplot(x=col, y='Price', data=df)
        plt.title(f'Scatter plot of {col} vs. Price')
plt.tight_layout()
plt.show()

# Document observations
print("Observations:")
print("The heatmap shows strong positive correlations between salary, partner salary, and total salary")
print("The boxplots and scatter plots reveal potential outliers.")
```
